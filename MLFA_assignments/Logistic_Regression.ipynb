```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00537/sobar-72.csv')
df
```

```python
X= df.iloc[:,0:19].values
Y= df.iloc[:,19].values
```

```python
# train the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 1/10, random_state = 42)
```

```python
#Sigmoid Function
#def sigmoid(Z):

def sigmoid(z):
  return 1.0/(1 + np.exp(-z))
```

```python
def loss(y, y_hat):
  loss = -np.mean(y*np.log(y_hat) - (1 - y)*np.log(1 - y_hat))
  return loss
```

```python
def gradients(X, y, y_hat):
    m = X.shape[0]
    dw = (1/m)*np.dot(X.T, (y_hat - y))
    db = (1/m)*np.sum((y_hat - y)) 
    return dw, db
```

```python
def normalize(X):
    m, n = X.shape
    for i in range(n):
        X = (X - X.mean(axis=0))/X.std(axis=0)       
    return X
```

```python
def train(X, y, bs, epochs, lr):
    m, n = X.shape
    w = np.random.uniform(size=(n,1))
    b = 0
    y = y.reshape(m,1)
    x = normalize(X)
    losses = []
    for epoch in range(epochs):
        for i in range((m-1)//bs + 1):
            start_i = i*bs
            end_i = start_i + bs
            xb = X[start_i:end_i]
            yb = y[start_i:end_i]
            y_hat = sigmoid(np.dot(xb, w) + b)
            dw, db = gradients(xb, yb, y_hat)
            w -= lr*dw
            b -= lr*db
        l = loss(y, sigmoid(np.dot(X, w) + b))
        losses.append(l)
    return w, b, losses
```

```python
def predict(X, w, b):
    x = normalize(X)
    preds = sigmoid(np.dot(X, w) + b)
    pred_class = []
    pred_class = [1 if i > 0.5 else 0 for i in preds]
    return np.array(pred_class)
```

```python
def accuracy(y, y_hat):
    accuracy = np.sum(y == y_hat) / len(y)
    return accuracy
```

```python
w, b, l = train(X_train, y_train, bs=1, epochs=1000, lr=0.0001)
print("The accuracy of the model is :",accuracy(y_test, y_hat=predict(X_test,w,b))*100,"%")
```
